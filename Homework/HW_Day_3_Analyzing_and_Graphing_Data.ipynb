```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# import data
url2="https://bulldog2.redlands.edu/fac/tamara_veenstra/ML/superbowl_stats.csv"
df=pd.read_csv(url2)
```

```python
# checked to make sure everything was imported properly
df
```

```python
# clean up the code 
df.drop(columns="Unnamed: 13", inplace=True)
df.drop([54,55,57], inplace=True)
df.dropna(inplace=True)
df.tail()
```

```python
# viewing the dataframe to make sure it all looks good
df
```

```python
# checking the datatypes
df.dtypes
```

```python
# changing the datatype of year to be an int
df["Year"]=pd.to_numeric(df["Year"],errors='coerce',downcast='integer')
```

```python
# confirming that the datatype of year was, in fact, changed
df.dtypes
```

```python
# setting an index
df.set_index('Year', inplace=True)
df.head()
```

```python
# changing the names in the columns
df.rename(columns={'1st':'Team1 1st', '2nd':'Team1 2nd', '3rd':'Team1 3rd', '4th':'Team1 4th', 'Final':'Team1 Final'}, inplace=True)
for col in df.columns:  ##loop through all the column names
    if col[-2:]==".1":  ## check if the end with .1
        new_name="Team2 "+col[0:-2] ## create a new name which is Team2 concatenated with the previous name minus the .1 part
        new_name_dict={col:new_name} ## dictionary has a single element in the form {old_name:new_name}
        df.rename(columns={col:new_name}, inplace=True) ## use rename for each column
df.head()
```

```python
# adding column for the final combined score
df['Combined final']=df['Team1 Final']+df['Team2 Final']
df.head()
```

```python
# showing just the last digit of the final score
df["Final Pool"]=df['Combined final']%10
df.head()
```

```python
# get a frequency count 
result=df['Final Pool'].value_counts()
result
```

```python
# get x and y values from result
type(result)
```

```python
print("column 1", result.index)
print("column 2", result. values)
```

```python
# Make a beautiful bar chart with the data!!
plt.bar(result.index,result.values,width=.8,color='purple')
plt.title("Frequency distribution for last digit of combined superbowl final scores")
plt.xlabel("Last digit")
plt.ylabel("Count")
plt.xticks(np.arange(10))
plt.yticks(np.arange(10))
plt.show()
```
